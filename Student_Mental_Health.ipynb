#**K-Nearest Neighbour**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE


# Membaca data dari file CSV
data = pd.read_csv('drive/My Drive/Data Analytics Project/Student Mental health.tsv', delimiter='\t', quoting=3)

# Membuat kolom baru 'Kategori Mental Health' berdasarkan nilai kolom-kolom yang ada
data['Kategori Mental Health'] = data.apply(lambda row:
    "tidak ada masalah" if row['Do you have Depression?'] == 'No' and row['Do you have Anxiety?'] == 'No' and row['Do you have Panic attack?'] == 'No'
    else "depresi berat" if row['Do you have Depression?'] == 'Yes' and row['Do you have Anxiety?'] == 'Yes' and row['Do you have Panic attack?'] == 'Yes'
    else "depresi sedang" if row['Do you have Depression?'] == 'Yes' and row['Do you have Anxiety?'] == 'Yes' and row['Do you have Panic attack?'] == 'No'
    else "depresi sedang" if row['Do you have Depression?'] == 'Yes' and row['Do you have Anxiety?'] == 'No' and row['Do you have Panic attack?'] == 'Yes'
    else "depresi sedang" if row['Do you have Depression?'] == 'No' and row['Do you have Anxiety?'] == 'Yes' and row['Do you have Panic attack?'] == 'Yes'
    else "depresi ringan" if row['Do you have Depression?'] == 'Yes' and row['Do you have Anxiety?'] == 'No' and row['Do you have Panic attack?'] == 'No'
    else "depresi ringan" if row['Do you have Depression?'] == 'No' and row['Do you have Anxiety?'] == 'Yes' and row['Do you have Panic attack?'] == 'No'
    else "depresi ringan" if row['Do you have Depression?'] == 'No' and row['Do you have Anxiety?'] == 'No' and row['Do you have Panic attack?'] == 'Yes'
    else "unknown", axis=1)

# Pembersihan data: Hapus baris dengan nilai non-numerik selain 'Year 1', 'Year 2', 'Year 3', dan 'Year 4' di kolom 'Your current year of Study'
valid_values = ['Year 1', 'Year 2', 'Year 3', 'Year 4']
data = data[data['Your current year of Study'].isin(valid_values)]

# Mengubah nilai 'Year 2' menjadi angka '2' dan lainnya dalam kolom 'Your current year of Study'
data.loc[data['Your current year of Study'] == 'Year 1', 'Your current year of Study'] = 1
data.loc[data['Your current year of Study'] == 'Year 2', 'Your current year of Study'] = 2
data.loc[data['Your current year of Study'] == 'Year 3', 'Your current year of Study'] = 3
data.loc[data['Your current year of Study'] == 'Year 4', 'Your current year of Study'] = 4

# Mengonversi range nilai CGPA menjadi nilai tengah dari setiap range
data['What is your CGPA?'] = data['What is your CGPA?'].apply(lambda x: (float(x.split('-')[0]) + float(x.split('-')[1])) / 2)

# Memilih atribut dan label
X = data[['Choose your gender', 'Age', 'What is your course?', 'Your current year of Study', 'What is your CGPA?', 'Marital status']]
y = data['Kategori Mental Health']

# Transformasi One-Hot Encoding untuk atribut-atribut kategorikal
categorical_features = ['Choose your gender', 'What is your course?', 'Marital status']
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough'
)


# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)

# Transformasi One-Hot Encoding
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)

# Normalisasi data (opsional)
scaler = StandardScaler(with_mean=False)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Apply SMOTE to balance the classes
smote = SMOTE(random_state=42,k_neighbors=2)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Buat model KNN dengan nilai K=3
knn = KNeighborsClassifier(n_neighbors=5,weights='distance', algorithm='auto',p=1)
knn.fit(X_train, y_train)

# Prediksi kelas untuk data uji
y_pred = knn.predict(X_test)

# Evaluasi model
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi Model: {accuracy}')

# Tampilkan laporan klasifikasi
print('Laporan Klasifikasi:')
print(classification_report(y_test, y_pred,zero_division=1))
```
